```python
# Simple Outlier Detection per Column (IQR Method)
import numpy as np
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

for col in df.select_dtypes(include=[np.number]).columns:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    
    print(f"\nColumn: {col}")
    print(f"Number of Outliers: {outliers.shape[0]}")
    print("Outlier Values:")
    print(outliers.values[:10])  # show first 10 outlier values


# For each numeric column:
# Q1 = 25th percentile (lower quartile).
# Q3 = 75th percentile (upper quartile).
# IQR = Q3 â€“ Q1 â†’ Interquartile Range (spread of middle 50% of the data).

# for col in df.select_dtypes(include=[np.number]).columns:
# Loops over all numeric columns only (ignores categorical/text).

# lower = Q1[col] - 1.5 * IQR[col]
# upper = Q3[col] + 1.5 * IQR[col]
# Defines the outlier cutoff range:
# Anything below Q1 â€“ 1.5 Ã— IQR is a low outlier.
# Anything above Q3 + 1.5 Ã— IQR is a high outlier.

# outliers = df[(df[col] < lower) | (df[col] > upper)][col]
# Filters the dataset and extracts all outlier values for that column.

# in short: 
# applies the IQR method to find values that are unusually high/low, 
# then prints how many outliers exist and shows a preview of them

```

```python
x = df.drop("churn", axis=1)   # independent variables
y = df["churn"]

# If categorical â†’ convert to numeric
x = pd.get_dummies(x, drop_first=True)
```

## Standard Scalar

```python
# Import libraries
from sklearn.preprocessing import StandardScaler

# Assuming X is your feature matrix and y is the target column
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

print("Before Scaling:\n", x.head())
print("\nAfter Scaling (first 5 rows):\n", x_scaled[:5])

# Different features in a dataset can have very different scales (e.g., minutes vs. charges vs. calls).

# StandardScaler standardizes all features so they have mean = 0 and standard deviation = 1.

# This ensures no feature dominates just because of its larger values.

# As a result, machine learning models train faster and give more accurate results.
```

## SMOTE ANALYSIS

```python
# Import library
from imblearn.over_sampling import SMOTE
from collections import Counter

# Before SMOTE
print("Before SMOTE:", Counter(y))

# Apply SMOTE
sm = SMOTE(random_state=42)
x_resampled, y_resampled = sm.fit_resample(x_scaled, y)

# After SMOTE
print("After SMOTE:", Counter(y_resampled))

#Before SMOTE: {0: 2850, 1: 483}
#0 = customers who did not churn â†’ 2850
#1 = customers who churned â†’ 483
#This is imbalanced (many more non-churners than churners).

#After SMOTE: {0: 2850, 1: 2850}
#Now, churners (1) were synthetically increased to match non-churners (0).
#Instead of just duplicating rows, SMOTE creates new synthetic samples of churners by 
#interpolating between existing churn data points.

#Why do we do this?
#Balance the dataset â†’ so the model doesnâ€™t get biased toward predicting â€œNo churnâ€.
# Improve accuracy for minority class â†’ the model will learn patterns of churners better.
# Fair training â†’ both churn and non-churn customers get equal importance.
```

## selecting best features using RFC

```python
#using random forest classifier algorithm we are selceting top 15 features which has most impact on the churn 
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

rf = RandomForestClassifier(random_state=42)
rf.fit(x, y)

# Get importance scores
importances = pd.Series(rf.feature_importances_, index=x.columns)

# Sort & show top 15 features
best_features = importances.sort_values(ascending=False).head(10)
print(best_features)

#as a result we get this 15 features printed `
```

```python
selected_columns = best_features.index.tolist()
print("Selected Independent Variables:", selected_columns)
```

## feature selection

```python
#based on the RFC best features selection we are selcting the the top 10 features for the independant variables
features=['total_charge', 'customer_service_calls', 'international_plan', 'day_mins', 'day_charge', 'voice_mail_messages', 'international_calls', 'voice_mail_plan', 'international_mins', 'evening_charge']
features
```

```python
#independant variables : so here we have selected top 10 features which has most impact on churn as an independand variables
x=df[features]
x
```

```python
#dependant variables (target variable) which we want to predict
y=df['churn']
y
```

## Train test split

```python
# 1) Train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=28)

# here during the train_test_split
#â†’ 80% of the data (used to train the model).
# â†’ 20% of the data (used to test/evaluate the model).

# ðŸ”¹ Parameters explained
# test_size=0.2 â†’ means 20% of the dataset goes to testing, 80% to training.
# random_state=28 â†’ ensures the split is the same every time you run it (reproducibility).
# Without random_state, youâ€™d get different train/test splits each time.

# Why we do this
# The model learns patterns from training data.
# Then we test it on unseen test data to check performance.
# This prevents overfitting (model memorizing instead of learning).

```

## model buiding

```python
#here in the model building we juz applied RFC algorithm
#RandomForestClassifier is an ensemble learning method: it builds many decision trees 
#and combines their predictions for better accuracy and stability.

#why random forest 
#it avoids overfitting
#it handles imbalanced data
#it selects the best features 
#handles non liner relation
#provides high accuracy
```

```python
# 2) Build RandomForestClassifier model
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest model
model = RandomForestClassifier(n_estimators=100,max_depth=None,random_state=42)    
# Fit the model
model.fit(x_train,y_train)

# This initializes a Random Forest model with three main parameters:
# n_estimators=100 â†’ The forest will have 100 decision trees. More trees usually improve performance .

# max_depth=None â†’ Trees are grown fully until leaves are pure or until each leaf 
#has less than min_samples_split samples. Setting a limit can prevent overfitting.

# random_state=42 â†’ Ensures reproducibility. 


# This trains the Random Forest model using  training data:
# x_train â†’ features (independent variables)
# y_train â†’ target variable (labels you want to predict)
# During .fit(), the model builds 100 decision trees, each trained on a random subset of data and features.
# After this, model learns patterns from the data so it can predict unseen samples.

```

## prediction

```python
# Make predictions on unseen dataset (test dataset)
y_pred = model.predict(x_test)
y_pred
```

## evaluation

```python
#evaluation : 
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
#out of 667 test datset 20% 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# so here we infer that 
# out of 667/564 are non churn customers
# out of 667/103 are churn customers
# so the accuarcy is 97 which is an excellent model
```

## DEPLOYMENT

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 28 01:02:39 2025

@author: Rajiv Anatwar
"""
import streamlit as st
import pandas as pd
import joblib

# --- This feature list MUST EXACTLY MATCH the training script ---
FEATURES_FOR_MODEL = [
    'total_charge', 'customer_service_calls', 'international_plan', 'day_mins', 
    'day_charge', 'voice_mail_messages', 'international_calls', 
    'voice_mail_plan', 'international_mins', 'evening_charge'
]

# --- 1. App Title ---
st.title("Customer Churn Predictor")
st.write("This app uses your custom XGBoost model to predict churn.")

# --- 2. Load Your Model ---
try:
    model = joblib.load('xgb_churn_model.joblib')
except FileNotFoundError:
    st.error("Error: 'xgb_churn_model.joblib' not found. Please run the training script first.")
    st.stop()

# --- 3. User Input Interface ---
# Create input fields ONLY for the 10 features your model needs
st.header("Enter Customer Details")

inputs = {}
inputs['day_mins'] = st.number_input('Total Day Minutes', min_value=0.0, format="%.2f")
inputs['day_charge'] = st.number_input('Total Day Charge ($)', min_value=0.0, format="%.2f")
inputs['evening_charge'] = st.number_input('Total Evening Charge ($)', min_value=0.0, format="%.2f")
inputs['international_mins'] = st.number_input('Total International Minutes', min_value=0.0, format="%.2f")
inputs['international_calls'] = st.number_input('Total International Calls', min_value=0)
inputs['total_charge'] = st.number_input('Total Charge ($)', min_value=0.0, format="%.2f")
inputs['customer_service_calls'] = st.number_input('Calls to Customer Service', min_value=0)
inputs['voice_mail_messages'] = st.number_input('Number of Voice Mail Messages', min_value=0)
international_plan_str = st.selectbox('Has International Plan?', ['no', 'yes'])
voice_mail_plan_str = st.selectbox('Has Voice Mail Plan?', ['no', 'yes'])

# --- 4. Prediction Logic ---
if st.button('Predict Churn'):
    # Convert string 'yes'/'no' to 1/0
    inputs['international_plan'] = 1 if international_plan_str == 'yes' else 0
    inputs['voice_mail_plan'] = 1 if voice_mail_plan_str == 'yes' else 0

    # Create the DataFrame and re-order it to match the training order
    input_df = pd.DataFrame([inputs])[FEATURES_FOR_MODEL]
    
    # Make the prediction
    prediction = model.predict(input_df)
    prediction_proba = model.predict_proba(input_df)
    
    # --- 5. Display the Prediction ---
    st.header("Prediction Result")
    if prediction[0] == 1:
        st.error(f"Prediction: LIKELY to Churn (Confidence: {prediction_proba[0][1]:.2%})")
    else:
        st.success(f"Prediction: UNLIKELY to Churn (Confidence: {prediction_proba[0][0]:.2%})")
```
